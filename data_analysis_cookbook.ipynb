## Plotting

```python
#Create Instance
PLT=PlottingMethods()
```

```python
#Display all plotting methods
response=PLT.get_methods_info()
pd.DataFrame(response.get('response',{}))
```

```python
inspector.df.columns
```

```python
inspector.df['count']=1
```

```python
result=PLT.plot_pie_chart(names='What do you expect MOST from this programme?',values='count',data=inspector.df)
result.get('status')
```

```python
PLT.display_image(result=result)
```

```python
result=PLT.plot_bar_chart(x='What do you expect MOST from this programme?',y='count',data=inspector.df)
```

```python
result=PLT.plot_histogram(x='What was your Z-score at A/L',data=inspector.df)
```

```python
title='Average GPA'
result=PLT.plot_heat_map(values='What is your GPA after 1st semester ?',index='What do you expect MOST from this programme?',columns='How do you rate your understanding of Maths & Physics?',aggregade_method='mean', title=title,data=inspector.df)
```

```python
result=PLT.plot_sankey_diagram(source_column='How do you rate your understanding of Maths & Physics?',target_column='How well do you understand what Mechanical Engineering involves?', values='count',data=inspector.df)
result.get('status')
```

```python
title='Maths and Physics vs Expected Destination'
result=PLT.plot_simple_sunburst_graph(path=['How do you rate your understanding of Maths & Physics?','At this stage, what are you thinking about after graduation?'], values='count', data=inspector.df,title=title)
result.get('status')
```
